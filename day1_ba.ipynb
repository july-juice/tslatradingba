```python
# Imports
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt

plt.style.use("ggplot")
```

```python
# Company Choice
ticker_symbol = "TSLA"
ticker = yf.Ticker(ticker_symbol)

print("Analysis on ", ticker_symbol)
```

```python
# Download financial data
financials = ticker.financials
financials
```

```python
# Extracting Revenue and Net Income
revenue = financials.loc["Total Revenue"]
net_income = financials.loc["Net Income"]

df = pd.DataFrame({
    "Revenue": revenue,
    "Net Income": net_income
})

df
```

```python
# KPI Calculations

rev_growth = df["Revenue"].pct_change()*100

profit_margin = (df["Net Income"]/ df["Revenue"]) * 100

kpi_table = pd.DataFrame({
    "Revenue": df["Revenue"],
    "Net Income": df["Net Income"],
    "Revenue Growth (%)": rev_growth,
    "Profit Margin (%)": profit_margin
})

kpi_table
```

```python
# Plotting revenue and net income
df.plot(kind="line", figsize=(10,5))
plt.title(f"{ticker_symbol} Revenue vs Net Income")
plt.ylabel("USD $")
plt.show()
```

```python
# Plot profit margin growth
profit_margin.plot(kind="line", figsize=(10,5))
plt.title(f"{ticker_symbol} Profit Margin (%)")
plt.ylabel("Margin (%)")
plt.show()
```
